# Python: Making Plots with Plotnine

This notebook contains material presented during Day 2 portion of Python. The official course materials are available here: https://datacarpentry.org/python-ecology-lesson/07-visualization-ggplot-python/index.html

`%matplotlib inline` = Jupyter-specific command when you want plots to appear in a Jupyter notebook

```python
%matplotlib inline
import plotnine as p9
import pandas as pd
```

```python
surveys_df = pd.read_csv('https://ndownloader.figshare.com/files/2292172')
surveys_complete = surveys_df.dropna().copy()
```

```python
surveys_complete
```

`ggplot()` creates a graphing environment and designates a data set to be used. Different components of the graph will be added to the base ggplot function using `+` symbols.

```python
(p9.ggplot(data=surveys_complete))
```

`aes()` = "aesthetic" mappings of data in a DataFrame to components of a graph. Tells ggplot that the data in the designated columns will be transformed to graphical elements of the graph. Aesthetic mappings alone do not produce data points on the graph.

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', 
                          y='hindfoot_length')))
```

**geom** = geometric representation of aesthetically-mapped data. These are the actual data points on the graph. Examples of geoms include `geom_point()` and `geom_bar()`.

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', 
                          y='hindfoot_length')) 
 + p9.geom_point())
```

The base ggplot declaration can be saved as a graph object. Additional components (geoms, xlab/ylab, etc.) can be added to this object.

```python
surveys_plot = (p9.ggplot(data=surveys_complete, 
                          mapping=p9.aes(x="weight", 
                                         y="hindfoot_length")))
```

```python
(surveys_plot 
 + p9.geom_point())
```

`alpha=` controls opacity. Here, alpha=0.1 sets opacity of points to 10%.

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', 
                          y='hindfoot_length'))
 + p9.geom_point(alpha=0.1))
```

`color=` controls the **outline** color of elements. Specify the fill color with `fill=`.

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length'))
 + p9.geom_point(alpha=0.1, color="blue"))
```

To color elements according to a variable, use `color='variable'` or `fill='variable'` in the aesthetic mappings. Plotnine will assign default colors for each unique element in that variable.

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', 
                          y='hindfoot_length', 
                          color='species_id'))
 + p9.geom_point(alpha=0.1))
```

`xlab()` and `ylab()` = custom axis titles

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', 
                          y='hindfoot_length', 
                          color='species_id'))
 + p9.geom_point(alpha=0.1)
 + p9.xlab("Weight (g)")
 + p9.ylab("Hindfood Length (mm)"))
```

**Scales** are transformations done on the data before converting them to visual elements. `scale_x_log10()` and `scale_y_log10()` = logarithmic scale for x and y

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', 
                          y='hindfoot_length', 
                          color='species_id'))
 + p9.geom_point(alpha=0.1)
 + p9.xlab("Weight (g)")
 + p9.ylab("Hindfood Length (mm)")
 + p9.scale_x_log10())
```

**Theme** controls additional visual elements of the graph that are not controlled by aesthetic mappings, geoms, or scales. `theme_bw()` is a default set of theme elements that produce a plain black and white figure. `theme()` allows for additional visual specifications.

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', 
                          y='hindfoot_length', 
                          color='species_id'))
 + p9.geom_point(alpha=0.1)
 + p9.xlab("Weight (g)")
 + p9.ylab("Hindfood Length (mm)")
 + p9.scale_x_log10()
 + p9.theme_bw()
 + p9.theme(text=p9.element_text(size=16)))
```

**Facets** = multiple identical graphs, each showing a different subset of the data according to a variable or variables specified. `facet_wrap('variable')` creates a graph for each unique element of a variable. `facet_grid('variable1 ~ variable2')` creates a grid of graphs showing how data vary with two specified variables. 

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight',
                          y='hindfoot_length',
                          color='species_id'))
 + p9.geom_point(alpha=0.1)
 + p9.facet_wrap('sex'))
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight',
                          y='hindfoot_length',
                          color='species_id'))
 + p9.geom_point(alpha=0.1)
 + p9.facet_wrap('plot_id'))
```
