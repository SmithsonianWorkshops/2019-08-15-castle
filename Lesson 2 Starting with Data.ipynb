## Filtering

```python
surveys_df['year'] == 2002
```

```python
surveys_df[surveys_df['year'] == 2002].head()
```

```python
year_2000 = surveys_df['year'] == 2000
species_dm = surveys_df['species_id'] == 'DM'

len(surveys_df[year_2000 & species_dm])
```

```python
len(surveys_df[year_2000 | species_dm])
```

```python
len(surveys_df[(surveys_df.year == 1999) & (surveys_df.weight <= 8.0)])
```

```python
target_species_ids = ['NL','DO','DS']
len(surveys_df[surveys_df['species_id'].isin(target_species_ids)])
```
